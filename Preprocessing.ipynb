### Feature engineering

```python
def get_day_period(time):
    hour = int(time.split(':')[0])
    period = ""
    if 6 <= hour <= 12:
        period = "manha"
    elif 13 <= hour <= 18:
        period = "tarde"
    elif 19 <= hour <= 23:
        period = "noite"
    elif 00 <= hour <= 5:
        period = "madrugada"
    return period

def get_month_period(day):
    period = ""
    if 1 <= day <= 10:
        period = "inicio"
    elif 11 <= day <= 20:
        period = "meio"
    elif 21 <= day <= 31:
        period = "fim"
    return period

def get_week_period(day):
    period = ""
    if day in range(0,5):
        period = "semana"
    else:
        period = "fds"

    return period

```

```python
full_df.loc[:, "dia"] = pd.to_datetime(full_df.loc[:, "dia_ocorrencia"]).dt.day.astype(int)
full_df.loc[:, "mes"] = pd.to_datetime(full_df.loc[:, "dia_ocorrencia"]).dt.month.astype(int)
full_df.loc[:, "ano"] = pd.to_datetime(full_df.loc[:, "dia_ocorrencia"]).dt.year.astype(int)
full_df.loc[:, "periodo_mes"] = full_df.loc[:, "dia"].apply(lambda x: get_month_period(x))
full_df.loc[:, "dia_semana"] = pd.to_datetime(full_df.loc[:, "dia_ocorrencia"]).dt.day_of_week.astype(int)
full_df.loc[:, "periodo_semana"] = full_df.loc[:, "dia_semana"].apply(lambda x: get_week_period(x))
full_df.loc[:, "hora"] = pd.to_datetime(full_df['horario'], format='%H:%M:%S').dt.hour
full_df.loc[:, "periodo_dia"] = full_df.loc[:, "horario"].apply(lambda x: get_day_period(x))
```
